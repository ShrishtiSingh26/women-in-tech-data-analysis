```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/Diversity in tech companies.csv")

# Columns to convert
cols_to_convert = ['Female %', 'Male %', '% White', '% Asian', '% Latino', '% Black', '% Multi', '% Other']

# Convert the columns to numeric, forcing errors to NaN
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Option 1: Drop rows with NaN values
df_cleaned = df.dropna(subset=cols_to_convert)

# Option 2: Alternatively, you could fill NaN values with the mean of each column (optional)
# df_cleaned = df.fillna(df.mean())

# Check the column types to ensure conversion was successful
print(df_cleaned[cols_to_convert].dtypes)

# --- Gender Diversity Over the Years ---
# Group by year and calculate the average Female and Male percentages
gender_trends = df_cleaned.groupby('Year')[['Female %', 'Male %']].mean()

# Plot gender diversity trends over the years
plt.figure(figsize=(10, 6))
sns.lineplot(data=gender_trends)
plt.title('Gender Diversity Over the Years')
plt.ylabel('Percentage')
plt.xlabel('Year')
plt.show()

# --- Ethnic Diversity Over the Years ---
# Group by year and calculate the average percentages for each ethnicity
ethnicity_trends = df_cleaned.groupby('Year')[['% White', '% Asian', '% Latino', '% Black', '% Multi', '% Other']].mean()

# Plot ethnic diversity trends over the years
plt.figure(figsize=(12, 8))
ethnicity_trends.plot(kind='line', figsize=(10, 6))
plt.title('Ethnic Diversity Over the Years')
plt.ylabel('Percentage')
plt.xlabel('Year')
plt.show()
# Correlation matrix for gender and ethnicity percentages
correlation_matrix = df[['Female %', 'Male %', '% White', '% Asian', '% Latino', '% Black', '% Multi', '% Other']].corr()

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Gender and Ethnic Diversity')
plt.show()


# --- Company-Level Analysis ---
# Group by company and calculate the average gender and ethnic percentages
company_diversity = df_cleaned.groupby('Company')[['Female %', 'Male %', '% White', '% Asian', '% Latino', '% Black', '% Multi', '% Other']].mean()

# Sort by highest female representation
company_diversity_sorted = company_diversity.sort_values(by='Female %', ascending=False)

# Display the top 10 companies with the highest female representation
print(company_diversity_sorted.head(10))
```
